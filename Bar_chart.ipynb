```python
import pandas as pd
import plotly.express as px

# Load your Excel file into a pandas DataFrame
df = pd.read_csv('/content/unicef_indicator_2.csv')

# Extracting required columns for the bar chart
df_chart = df[['country', 'time_period', 'obs_value']]

# Assuming 'country' and 'time_period' are categorical variables and 'obs_value' is numeric
# Create a bar chart using plotly express
fig = px.bar(df_chart, x='country', y='obs_value', color='time_period',
             title='Proportion of population using improved drinking water sources over time')

# Update the layout to adjust figure properties
fig.update_layout(xaxis_title='Country', yaxis_title='Observed Value')

# Show the plot
fig.show()
```
