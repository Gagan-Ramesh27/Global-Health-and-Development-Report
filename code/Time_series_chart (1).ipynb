```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
import numpy as np
df = pd.read_csv('/content/unicef_metadata.csv')
df_cleaned = df.dropna(subset=['Life expectancy at birth total (years)'])
df_plot = df_cleaned[['country', 'year', 'Population total', 'Life expectancy at birth total (years)']]
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Population total', y='Life expectancy at birth total (years)', data=df_plot, hue='year', palette='viridis')
X = df_plot['Population total'].values.reshape(-1, 1)
y = df_plot['Life expectancy at birth total (years)'].values.reshape(-1, 1)
model = LinearRegression()
model.fit(X, y)
x_values = np.linspace(df_plot['Population total'].min(), df_plot['Population total'].max(), 100)
y_values = model.predict(x_values.reshape(-1, 1))
plt.plot(x_values, y_values, linestyle='--', color='red', linewidth=2, label='Linear Regression')
plt.title('Population vs Life Expectancy')
plt.xlabel('Population total')
plt.ylabel('Life expectancy at birth total (years)')
plt.legend()
plt.show()
```
