```python
import pandas as pd
df = pd.read_csv('/content/unicef_indicator_1.csv')
df_cleaned = df[['country', 'time_period', 'obs_value']].copy()
df_cleaned['time_period'] = pd.to_datetime(df_cleaned['time_period'], format='%Y')
df_cleaned['obs_value'] = df_cleaned['obs_value'].astype(float)
```

```python
import plotly.express as px
fig = px.choropleth(
    df_cleaned,
    locations='country',
    locationmode='country names',
    color='obs_value',
    hover_name='country',
    animation_frame='time_period',
    color_continuous_scale='Viridis',
    title='Proportion of population using of improved drinking water sources free from faecal and priority chemical contamination'
)
fig.update_layout(
    geo=dict(showframe=False, showcoastlines=False),
    coloraxis_colorbar=dict(title='Observation Value'),
    title_x=0.5
)
fig.show()
```

```python
fig = px.bar(df_cleaned, x='country', y='obs_value', color='time_period',
             title='Proportion of population using improved drinking water sources over time')

fig.update_layout(xaxis_title='Country', yaxis_title='Observed Value')
fig.show()
```
