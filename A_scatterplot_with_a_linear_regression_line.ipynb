```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv('/content/unicef_metadata.csv')

# Clean the data
# Convert 'Life expectancy at birth total (years)' column to numeric
df['Life expectancy at birth total (years)'] = pd.to_numeric(df['Life expectancy at birth total (years)'], errors='coerce')

# Drop rows with missing values in 'Life expectancy at birth total (years)' column
df.dropna(subset=['Life expectancy at birth total (years)'], inplace=True)

# Create a scatter plot with regression line
sns.set(style="whitegrid")  # Set style
plt.figure(figsize=(10, 6))  # Set figure size

# Plot scatter plot
sns.scatterplot(x='year', y='Life expectancy at birth total (years)', data=df)

# Add linear regression line
sns.regplot(x='year', y='Life expectancy at birth total (years)', data=df, scatter=False)

# Set plot labels and title
plt.xlabel('Year')
plt.ylabel('Life expectancy at birth (years)')
plt.title('Life Expectancy Over Time')

# Show plot
plt.show()
```
